#### The source dataset does not provide geographic metadata, so proxy data was used to demonstrate the multi-source integration pipeline

```python
# Creating proxy historical weather data
print("Start Date:", df["DateTime"].min())
print("End Date:", df["DateTime"].max())
```

```python
# creating an hourly weather timeline
weather_datetime = pd.date_range(
    start=df["DateTime"].min(),
    end=df["DateTime"].max(),
    freq="h"
)

weather_df = pd.DataFrame({
    "DateTime": weather_datetime
})

weather_df.head()
```

#### Creating weather variables

```python
# creating proxy values for pipeline development
# Using a fixed random seed so your results remain reproducible
np.random.seed(42)
```

```python
weather_df["Temperature"] = np.random.normal(
    loc=25,
    scale=5,
    size=len(weather_df)
)

weather_df["Humidity"] = np.random.normal(
    loc=65,
    scale=15,
    size=len(weather_df)
)

weather_df["WindSpeed"] = np.random.normal(
    loc=12,
    scale=5,
    size=len(weather_df)
)

weather_df["Precipitation"] = np.random.exponential(
    scale=1.5,
    size=len(weather_df)
)
```

```python
weather_df.head()
```

```python
# correcting inconsistencies
weather_df["Humidity"] = weather_df["Humidity"].clip(
    lower=0,
    upper=100
)

weather_df["WindSpeed"] = weather_df["WindSpeed"].clip(
    lower=0
)

weather_df["Precipitation"] = weather_df["Precipitation"].clip(
    lower=0
)
```

```python
weather_df["Temperature"] = weather_df["Temperature"].round(2)
weather_df["Humidity"] = weather_df["Humidity"].round(2)
weather_df["WindSpeed"] = weather_df["WindSpeed"].round(2)
weather_df["Precipitation"] = weather_df["Precipitation"].round(2)
```

```python
weather_df.head()
```

#### To demonstrate the cleaning pipeline, we can introduce a small number of missing values only into the proxy weather data

```python
missing_indices = np.random.choice(
    weather_df.index,
    size=100,
    replace=False
)

weather_df.loc[
    missing_indices,
    "Temperature"
] = np.nan
```

```python
weather_df.isnull().sum()
```

```python
# Handling missing weather values
weather_df["Temperature"] = (
    weather_df["Temperature"]
    .interpolate(method="linear")
)
```

#### Creating the event dataset

```python
# creating all unique dates
event_df = pd.DataFrame({
    "Date": pd.date_range(
        start=df["DateTime"].min().date(),
        end=df["DateTime"].max().date(),
        freq="D"
    )
})
```

```python
event_df.head()
```

```python
# Creating event features
np.random.seed(42)

event_df["SportsEvent"] = np.random.choice(
    [0, 1],
    size=len(event_df),
    p=[0.95, 0.05]
)

event_df["Concert"] = np.random.choice(
    [0, 1],
    size=len(event_df),
    p=[0.97, 0.03]
)

event_df["PublicDemonstration"] = np.random.choice(
    [0, 1],
    size=len(event_df),
    p=[0.99, 0.01]
)
```

```python
event_df["IsHoliday"] = np.nan
```

```python
event_df["HasEvent"] = np.where(
    (
        (event_df["SportsEvent"] == 1) |
        (event_df["Concert"] == 1) |
        (event_df["PublicDemonstration"] == 1)
    ),
    1,
    0
)
```

```python
event_df.head()
```

#### Prepare the data types before merging

```python
df["DateTime"] = pd.to_datetime(df["DateTime"])
weather_df["DateTime"] = pd.to_datetime(
    weather_df["DateTime"]
)

df["Date"] = pd.to_datetime(df["Date"])
event_df["Date"] = pd.to_datetime(
    event_df["Date"]
)
```

```python
print(df["DateTime"].dtype)
print(weather_df["DateTime"].dtype)

print(df["Date"].dtype)
print(event_df["Date"].dtype)
```

#### Merge traffic and weather data

```python
merged_df = pd.merge(
    df,
    weather_df,
    on="DateTime",
    how="left"
)
```

```python
merged_df.head()
```

```python
print("Original Traffic Rows:", len(df))
print("After Weather Merge:", len(merged_df))
```

```python
merged_df.isnull().sum()
```

#### Merge event data

```python
merged_df = pd.merge(
    merged_df,
    event_df,
    on="Date",
    how="left"
)
```
